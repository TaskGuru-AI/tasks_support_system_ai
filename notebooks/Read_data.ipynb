```python
import pandas as pd
import re
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('tickets.csv', sep=';', encoding='windows-1251', header = None, usecols=[i for i in range(5)])
df.columns=['start_dt','end_dt', 'sender_name', 'sender_email', 'theme']

# считываем первые пять корректных полей в датафрейм
```

```python
df.head()
```

```python
df2 = pd.read_csv('tickets.csv',
                 sep=';',
                 header = None,
                 encoding='windows-1251')

# в отдельный датафрейм считываем все колонки
```

```python
df2['descr'] = df2.iloc[:, 5:].apply(lambda row: ';'.join(row.dropna().astype(str)), axis=1)

# объединяем данные с 6 колонки до последней в одну
```

```python
df_fin = pd.concat([df, df2['descr']], axis=1)
```

```python
df_fin.iloc[:5, 5] #проверка 
```

```python
df_fin['descr'] = df_fin['descr'].str.replace('&quot;', '"')
df_fin['descr'].head()
# заменяем &quot на кавычки
```

```python
df_fin['descr'] = df_fin['descr'].str.replace(r'\s{2,}', ' ', regex=True)
#убираем лишние пробельные символы
```

```python
df_fin.to_csv('tickets_2.csv', sep=';', index=False, encoding='windows-1251')
```
