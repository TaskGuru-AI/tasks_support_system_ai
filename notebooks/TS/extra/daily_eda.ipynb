```python
import pandas as pd
import matplotlib.pyplot as plt
from tasks_support_system_ai.utils import get_correct_data_path
```

```python
df = pd.read_csv(get_correct_data_path("tickets_daily/tickets_daily.csv"), sep=";")
df["date"] = pd.to_datetime(df["date"], format="%d.%m.%Y")
```

```python
daily_sum = df.groupby("date")["new_tickets"].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(daily_sum["date"], daily_sum["new_tickets"], marker="o")
plt.title("Sum of Daily Tickets Across All Queues")
plt.xlabel("Date")
plt.ylabel("Number of Tickets")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
unique_queues = df.groupby("date")["queueId"].nunique().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(unique_queues["date"], unique_queues["queueId"], marker="o")
plt.title("Number of Unique Queues per Day")
plt.xlabel("Date")
plt.ylabel("Number of Unique Queues")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
